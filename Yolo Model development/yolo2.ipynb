```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import os

BASE_DIR = "/content/drive/MyDrive/Syncrones/yolo"   # <<< CHANGE HERE
DATA_DIR = BASE_DIR                                   # images/ and labels/ live directly here

print("BASE_DIR:", BASE_DIR)

train_img_dir = f"{DATA_DIR}/images/train"
val_img_dir   = f"{DATA_DIR}/images/val"

print("\nTrain dir exists?", os.path.exists(train_img_dir))
print("Val dir exists?", os.path.exists(val_img_dir))

if os.path.exists(train_img_dir):
    print("Train images:", os.listdir(train_img_dir))
```

```python
import os

# image folders
os.makedirs(f"{DATA_DIR}/images/train", exist_ok=True)
os.makedirs(f"{DATA_DIR}/images/val", exist_ok=True)

# label folders
os.makedirs(f"{DATA_DIR}/labels/train", exist_ok=True)
os.makedirs(f"{DATA_DIR}/labels/val", exist_ok=True)

print(os.listdir(f"{DATA_DIR}/images/train"))
```

```python
from glob import glob
from PIL import Image

CLASS_ID = 0  # 0 = text_block

for split in ["train", "val"]:
    img_dir = os.path.join(DATA_DIR, "images", split)
    lab_dir = os.path.join(DATA_DIR, "labels", split)
    os.makedirs(lab_dir, exist_ok=True)

    img_paths = glob(os.path.join(img_dir, "*"))
    print(f"{split}: found {len(img_paths)} images")

    for img_path in img_paths:
        img = Image.open(img_path)
        w, h = img.size

        # one big box (90% of image, centered)
        x_center = 0.5
        y_center = 0.5
        box_w    = 0.9
        box_h    = 0.9

        yolo_line = f"{CLASS_ID} {x_center} {y_center} {box_w} {box_h}\n"

        stem = os.path.splitext(os.path.basename(img_path))[0]
        label_path = os.path.join(lab_dir, stem + ".txt")

        with open(label_path, "w") as f:
            f.write(yolo_line)

print("Dummy label files created.")
```
