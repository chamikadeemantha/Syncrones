Train YOLO on your layout data

```python
# Load a small pre-trained YOLO model
model = YOLO("yolov8n.pt")   # n = nano, smallest

data_yaml = "/content/drive/MyDrive/Syncrones/yolo/layout.yaml"

results = model.train(
    data=data_yaml,
    epochs=50,          # you can reduce to 10â€“20 for testing
    imgsz=640,
    batch=4,
    name="layout_yolo",
    project="/content/drive/MyDrive/Syncrones/yolo/runs"
)
```

Run inference on a page

```python
from ultralytics import YOLO

# Your exact model path
weights_path = "/content/drive/MyDrive/Syncrones/yolo/runs/layout_yolo7/weights/best.pt"
model = YOLO(weights_path)

# Your test image
img_path = "/content/drive/MyDrive/Syncrones/yolo/datasets/layout/images/train/1.png"

results = model.predict(
    source=img_path,
    imgsz=640,
    conf=0.25,
    save=True,
    project="/content/drive/MyDrive/Syncrones/yolo/results",
    name="layout_infer"
)

print("Done. Check /content/drive/MyDrive/Syncrones/yolo/results/layout_infer")
```
